```python
import random

print("Bem-vindo ao jogo da adivinhação!")
print("Escolha o nível de dificuldade:")
print("1 - Fácil (1 a 50)")
print("2 - Médio (1 a 100)")
print("3 - Difícil (1 a 200)")

nivel = int(input("Digite o número da dificuldade: "))

# Definindo intervalo conforme dificuldade
if nivel == 1:
    limite_superior = 50
elif nivel == 2:
    limite_superior = 100
else:
    limite_superior = 200

numero_secreto = random.randint(1, limite_superior)
tentativas = 0
limite_tentativas = 10

print(f"Tente adivinhar o número entre 1 e {limite_superior}. Você tem {limite_tentativas} tentativas.")

acertou = False

while tentativas < limite_tentativas and not acertou:
    palpite = int(input("Digite seu palpite: "))
    tentativas += 1

    if palpite == numero_secreto:
        print(f"Parabéns! Você acertou em {tentativas} tentativas.")
        acertou = True
    elif palpite < numero_secreto:
        print("O número secreto é maior.")
    else:
        print("O número secreto é menor.")

if not acertou:
    print(f"Fim de jogo! O número secreto era {numero_secreto}.")
```
