```python
a = float(input("Digite o primeiro número:"))
b = float(input("Digite o segundo número:"))
op = input("Digite a operação (+, -, *, /): ")

if op == "+":
  print("Resultado:", a + b)
elif op == "-":
  print("Resultado:", a - b)
elif op == "*":
  print("Resultado:", a * b)
elif op == "/":
  print("Resultado:", a / b)
else:
  print("Operação inválida")
```
